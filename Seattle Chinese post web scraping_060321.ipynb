```python
from textblob import TextBlob
```

```python
chinese= []
for li in soup.find_all('h2'):
    chinese.append(li.text)

Chinesestr = ''.join(chinese)
blob = TextBlob(Chinesestr)
print(Chinesestr)
Englishstr = blob.translate(to="EN")
print('------------------------------------------------')
print(Englishstr)
```

```python
#Create a List of Headlines in the variable 'lead'
lead =[]
headlines =soup.find_all('h2', class_ = 'entry-title')
for headline in headlines:
    if len(headline)>0: #Eliminates empty strings
        lead.append(headline.text)
lead
```

```python
#make sure the text is in a list
# create a new string = ''.join(name of list)
Lead_string = ''.join(lead)
#make a blob using textblob
blob = TextBlob(Lead_string)
#translate to English
englishstr_lead = blob.translate(to="EN")
print(englishstr_lead)
```

```python
#Create a List of news summaries in the variable 'news_summaries'
news_summaries =[]
summaries =soup.find_all('div', class_='entry-content')
for summary in summaries:
    news_summaries.append(summary.text)
news_summaries
```

```python
#make sure the text is in a list
# create a new string = ''.join(name of list)
news_summaries_string = ''.join(news_summaries)
#make a blob using textblob
blob = TextBlob(news_summaries_string)
#translate to English
englishstr_summaries = blob.translate(to="EN")
print(englishstr_summaries)
```

```python
#Create a List of website source in the variable 'sources'
sources =[]
websites =soup.find_all('span', class_='entry-author-name')
for website in websites:
    sources.append(website.text)
sources
```

```python
#Create a List of news summaries in the variable 'news_summaries'
dates =[]
article_dates =soup.find_all('time', class_ = 'entry-time')
for date in article_dates:
    dates.append(date.text)
dates
```
